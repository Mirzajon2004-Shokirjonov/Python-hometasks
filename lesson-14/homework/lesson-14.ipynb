```python
import json
import os

FILE_PATH = "C:\\Users\\ITPARK\\Downloads\\books.json"

def initialize_file():
    if not os.path.exists(FILE_PATH):
        with open(FILE_PATH, "w") as f:
            json.dump([], f)

def read_books():
    with open(FILE_PATH, "r") as f:
        return json.load(f)

def write_books(books):
    with open(FILE_PATH, "w") as f:
        json.dump(books, f, indent=4)

def add_book():
    books = read_books()
    new_id = max([book["id"] for book in books], default=0) + 1
    title = input("Kitob nomi: ")
    author = input("Muallif: ")
    year = int(input("Yili: "))
    books.append({"id": new_id, "title": title, "author": author, "year": year})
    write_books(books)
    print("Kitob qo'shildi.")

def update_book():
    books = read_books()
    book_id = int(input("Yangilamoqchi bo'lgan kitob ID raqami: "))
    for book in books:
        if book["id"] == book_id:
            book["title"] = input(f"Yangi nom ({book['title']}): ") or book["title"]
            book["author"] = input(f"Yangi muallif ({book['author']}): ") or book["author"]
            book["year"] = int(input(f"Yangi yil ({book['year']}): ") or book["year"])
            write_books(books)
            print("Kitob yangilandi.")
            return
    print("Kitob topilmadi.")

def delete_book():
    books = read_books()
    book_id = int(input("O'chirmoqchi bo'lgan kitob ID raqami: "))
    new_books = [book for book in books if book["id"] != book_id]
    if len(books) == len(new_books):
        print("Kitob topilmadi.")
    else:
        write_books(new_books)
        print("Kitob o'chirildi.")

def list_books():
    books = read_books()
    if not books:
        print("Kitoblar ro'yxati bo'sh.")
    else:
        for book in books:
            print(f"ID: {book['id']} | Nomi: {book['title']} | Muallif: {book['author']} | Yil: {book['year']}")

def main():
    initialize_file()
    while True:
        print("\nKitoblar Boshqaruvi")
        print("1. Kitob qo'shish")
        print("2. Kitobni yangilash")
        print("3. Kitobni o'chirish")
        print("4. Barcha kitoblarni ko'rish")
        print("5. Chiqish")
        choice = input("Tanlang (1-5): ")

        if choice == '1':
            add_book()
        elif choice == '2':
            update_book()
        elif choice == '3':
            delete_book()
        elif choice == '4':
            list_books()
        elif choice == '5':
            break
        else:
            print("Noto'g'ri tanlov.")

if __name__ == "__main__":
    main()
```

```python

import requests
import random
API_KEY = '1165ec4e'


movie_db = {
    "action": ["Mad Max: Fury Road", "John Wick", "Gladiator"],
    "comedy": ["The Mask", "Superbad", "The Hangover"],
    "drama": ["Forrest Gump", "The Shawshank Redemption", "Fight Club"],
    "horror": ["Get Out", "A Quiet Place", "The Conjuring"],
    "sci-fi": ["Inception", "Interstellar", "The Matrix"]
}

def recommend_movie(genre):
    genre = genre.lower()
    if genre not in movie_db:
        print("Kechirasiz, bu janr mavjud emas.")
        return
    
    movie_title = random.choice(movie_db[genre])
    print(f"Tavsiya etilgan film: {movie_title}")

    
    response = requests.get(f"http://www.omdbapi.com/?apikey={API_KEY}&t={movie_title}")
    
    if response.status_code == 200:
        data = response.json()
        if data["Response"] == "True":
            print("\nFilm haqida ma'lumot:")
            print(f"Nomi: {data['Title']}")
            print(f"Yil: {data['Year']}")
            print(f"Janr: {data['Genre']}")
            print(f"IMDB Reytingi: {data['imdbRating']}")
            print(f"Qisqacha: {data['Plot']}")
        else:
            print("Film topilmadi.")
    else:
        print("Ma'lumot olishda xatolik yuz berdi.")

if __name__ == "__main__":
    user_genre = input("Janrni tanlang (action, comedy, drama, horror, sci-fi): ")
    recommend_movie(user_genre)
```

```python
import requests
import random
API_KEY = '1165ec4e'
BASE_URL = "http://www.omdbapi.com/"


def recommend_movie_by_genre(genre):
    
    response = requests.get(BASE_URL, params={
        'apikey': API_KEY,
        's': genre,
        'type': 'movie'
    })

    if response.status_code == 200:
        data = response.json()
        
        
        if data.get('Response') == 'True':
            movies = data.get('Search', [])
            
            
            if movies:
                random_movie = random.choice(movies)
                movie_title = random_movie['Title']
                movie_year = random_movie['Year']
                movie_imdb_id = random_movie['imdbID']
                print(f"Tavsiya etilgan film: {movie_title} ({movie_year})")
                print(f"IMDB ID: {movie_imdb_id}")
            else:
                print("Ushbu janrga mos film topilmadi.")
        else:
            print("Filmlar ma'lumotlari topilmadi.")
    else:
        print("API so'rov xatosi.")

genre = input("Film janrini kiriting (masalan: Action, Comedy, Drama): ")
recommend_movie_by_genre(genre)
```
