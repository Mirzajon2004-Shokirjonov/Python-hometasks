```python
class Task:
    def __init__(self, title, description, due_date):
        self.title = title
        self.description = description
        self.due_date = due_date
        self.status = "incomplete"

    def mark_complete(self):
        self.status = "complete"

    def __str__(self):
        return f"Title: {self.title}\nDescription: {self.description}\nDue Date: {self.due_date}\nStatus: {self.status}\n"


class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def mark_task_complete(self, index):
        if 0 <= index < len(self.tasks):
            self.tasks[index].mark_complete()
            print(" Task marked as complete.")
        else:
            print(" Noto'g'ri indeks.")

    def list_all_tasks(self):
        if not self.tasks:
            print("Hech qanday topshiriq mavjud emas.")
            return
        for i, task in enumerate(self.tasks):
            print(f"Task #{i}")
            print(task)

    def list_incomplete_tasks(self):
        incomplete_tasks = [task for task in self.tasks if task.status == "incomplete"]
        if not incomplete_tasks:
            print(" Barcha topshiriqlar bajarilgan.")
            return
        for i, task in enumerate(incomplete_tasks):
            print(f"Incomplete Task #{i}")
            print(task)



def main():
    todo = ToDoList()

    while True:
        print("\n ToDo List Menu:")
        print("1. Add new task")
        print("2. Mark task as complete")
        print("3. List all tasks")
        print("4. List incomplete tasks")
        print("5. Exit")

        choice = input("Tanlovingizni kiriting (1-5): ")

        if choice == '1':
            title = input("Sarlavha: ")
            description = input("Tavsif: ")
            due_date = input("Muddati (YYYY-MM-DD): ")
            task = Task(title, description, due_date)
            todo.add_task(task)
            print(" Topshiriq qo'shildi.")
        elif choice == '2':
            todo.list_all_tasks()
            try:
                index = int(input("Bajarilgan topshiriq raqamini kiriting: "))
                todo.mark_task_complete(index)
            except ValueError:
                print(" Noto'g'ri raqam.")
        elif choice == '3':
            todo.list_all_tasks()
        elif choice == '4':
            todo.list_incomplete_tasks()
        elif choice == '5':
            print(" Dasturdan chiqilmoqda...")
            break
        else:
            print(" Noto'g'ri tanlov.")


if __name__ == "__main__":
    main()
```

```python
class Post:
    def __init__(self, title, content, author):
        self.title = title
        self.content = content
        self.author = author

    def __str__(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nContent: {self.content}\n"
class Blog:
    def __init__(self):
        self.posts = []

    def add_post(self, post):
        self.posts.append(post)
        print("Post qo‘shildi.")

    def list_all_posts(self):
        if not self.posts:
            print("Hech qanday post mavjud emas.")
        for i, post in enumerate(self.posts):
            print(f"Post #{i}")
            print(post)

    def posts_by_author(self, author):
        filtered = [post for post in self.posts if post.author.lower() == author.lower()]
        if not filtered:
            print(f"'{author}' muallifga tegishli hech qanday post topilmadi.")
        else:
            for post in filtered:
                print(post)

    def delete_post(self, index):
        if 0 <= index < len(self.posts):
            deleted = self.posts.pop(index)
            print(f" Post '{deleted.title}' o'chirildi.")
        else:
            print(" Noto'g'ri indeks.")

    def edit_post(self, index, new_title, new_content):
        if 0 <= index < len(self.posts):
            self.posts[index].title = new_title
            self.posts[index].content = new_content
            print(" Post yangilandi.")
        else:
            print(" Notogri indeks.")

    def latest_posts(self, count=3):
        if not self.posts:
            print(" Hech qanday post mavjud emas.")
            return
        print(f"So'nggi {count} ta post:")
        for post in self.posts[-count:][::-1]:  
            print(post)

def main():
    blog = Blog()

    while True:
        print("\n Simple Blog Menu:")
        print("1. Add new post")
        print("2. List all posts")
        print("3. Show posts by author")
        print("4. Delete a post")
        print("5. Edit a post")
        print("6. Show latest posts")
        print("7. Exit")

        choice = input("Tanlovingizni kiriting (1-7): ")

        if choice == '1':
            title = input("Sarlavha: ")
            content = input("Kontent: ")
            author = input("Muallif: ")
            blog.add_post(Post(title, content, author))

        elif choice == '2':
            blog.list_all_posts()

        elif choice == '3':
            author = input("Muallif ismini kiriting: ")
            blog.posts_by_author(author)

        elif choice == '4':
            blog.list_all_posts()
            try:
                index = int(input("O'chirmoqchi bo'lgan post raqamini kiriting: "))
                blog.delete_post(index)
            except ValueError:
                print("Raqam noto'g'ri.")

        elif choice == '5':
            blog.list_all_posts()
            try:
                index = int(input("Tahrir qilmoqchi bo'lgan post raqamini kiriting: "))
                new_title = input("Yangi sarlavha: ")
                new_content = input("Yangi kontent: ")
                blog.edit_post(index, new_title, new_content)
            except ValueError:
                print("Raqam noto'g'ri.")

        elif choice == '6':
            try:
                count = int(input("Nechta so'nggi post ko'rilsin? (default 3): ") or 3)
                blog.latest_posts(count)
            except ValueError:
                print(" Raqam noto'g'ri.")

        elif choice == '7':
            print("Dasturdan chiqilmoqda...")
            break
        else:
            print(" Noto'g'ri tanlov.")
if __name__ == "__main__":
    main()
```

```python
class Account:
    def __init__(self, account_number, holder_name, balance=0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} so'm depozit qilindi.")
        else:
            print("Noto'g'ri summa.")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount} so'm yechildi.")
        else:
            print("Hisobda mablag' yetarli emas.")

    def transfer(self, target_account, amount):
        if amount <= self.balance:
            self.balance -= amount
            target_account.balance += amount
            print(f"{amount} so'm {target_account.holder_name} ga o'tkazildi.")
        else:
            print("Hisobda mablag' yetarli emas.")

    def display(self):
        print(f"Hisob raqami: {self.account_number}")
        print(f"Egasi: {self.holder_name}")
        print(f"Balans: {self.balance} so'm\n")


class Bank:
    def __init__(self):
        self.accounts = {}

    def add_account(self, account):
        if account.account_number in self.accounts:
            print("Bu hisob raqami allaqachon mavjud.")
        else:
            self.accounts[account.account_number] = account
            print("Hisob muvaffaqiyatli yaratildi.")

    def get_account(self, acc_num):
        return self.accounts.get(acc_num, None)

    def display_all_accounts(self):
        if not self.accounts:
            print("Hech qanday hisob mavjud emas.")
        else:
            for acc in self.accounts.values():
                acc.display()


def main():
    bank = Bank()

    while True:
        print("\nSimple Banking System Menu:")
        print("1. Yangi hisob ochish")
        print("2. Balansni tekshirish")
        print("3. Pul qo'yish")
        print("4. Pul yechish")
        print("5. Pul o'tkazish")
        print("6. Barcha hisoblarni ko'rish")
        print("7. Chiqish")

        choice = input("Tanlovingizni kiriting (1-7): ")

        if choice == '1':
            acc_num = input("Hisob raqami: ")
            holder = input("Hisob egasi ismi: ")
            try:
                bal = float(input("Boshlang'ich balans: "))
                acc = Account(acc_num, holder, bal)
                bank.add_account(acc)
            except ValueError:
                print("Noto'g'ri miqdor.")

        elif choice == '2':
            acc_num = input("Hisob raqamini kiriting: ")
            acc = bank.get_account(acc_num)
            if acc:
                acc.display()
            else:
                print("Hisob topilmadi.")

        elif choice == '3':
            acc_num = input("Hisob raqamini kiriting: ")
            acc = bank.get_account(acc_num)
            if acc:
                try:
                    amount = float(input("Summani kiriting: "))
                    acc.deposit(amount)
                except ValueError:
                    print("Noto'g'ri summa.")
            else:
                print("Hisob topilmadi.")

        elif choice == '4':
            acc_num = input("Hisob raqamini kiriting: ")
            acc = bank.get_account(acc_num)
            if acc:
                try:
                    amount = float(input("Yechiladigan summani kiriting: "))
                    acc.withdraw(amount)
                except ValueError:
                    print("Noto'g'ri summa.")
            else:
                print("Hisob topilmadi.")

        elif choice == '5':
            from_acc = input("Yuboruvchi hisob raqami: ")
            to_acc = input("Qabul qiluvchi hisob raqami: ")
            sender = bank.get_account(from_acc)
            receiver = bank.get_account(to_acc)
            if sender and receiver:
                try:
                    amount = float(input("O'tkaziladigan summa: "))
                    sender.transfer(receiver, amount)
                except ValueError:
                    print("Noto'g'ri summa.")
            else:
                print("Biror hisob topilmadi.")

        elif choice == '6':
            bank.display_all_accounts()

        elif choice == '7':
            print("Dasturdan chiqilmoqda...")
            break

        else:
            print("Noto'g'ri tanlov.")


if __name__ == "__main__":
    main()
```
